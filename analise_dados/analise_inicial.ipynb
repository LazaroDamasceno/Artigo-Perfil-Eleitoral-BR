```python
import dask.dataframe as dd
import polars as pl
import polars.selectors as ps
```

# USANDO DASK

```python
df_dd = dd.read_csv(
    './dados/*.csv',
    encoding='latin',
    assume_missing=True,
    dtype={
        'RELIGIAO': 'object',
        'VINCULACAO_ANO': 'object'
    }
)
```

```python
df_dd.head()
```

```python
df_dd.columns
```

```python
df_dd = df_dd.drop(columns={
    'ANO_NASCIMENTO',
    'PESO',
    'ALTURA',
    'CABECA', 
    'CALCADO', 
    'CINTURA',
    'RELIGIAO',
    'JSM',
    'MUN_JSM', 
    'UF_JSM',
    'MUN_NASCIMENTO',
    'UF_NASCIMENTO',
    'PAIS_NASCIMENTO',
    'MUN_RESIDENCIA',
    'UF_RESIDENCIA',
    'PAIS_RESIDENCIA',
    'ZONA_RESIDENCIAL'
})
```

```python
df_dd.dtypes
```

```python
dd.to_parquet(
    df_dd, 
    './dados_parquet/', 
    engine='pyarrow', 
    write_index=False
)
```
